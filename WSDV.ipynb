```python
#Problem 1

import requests
from bs4 import BeautifulSoup
url = "https://en.wikipedia.org/wiki/Mary_Oliver"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
links = soup.find_all("a")
for link in links:
    href = link.get("href")
    if href:
        print(href)

mary_oliver = []
for link in links:
    href = link.get("href")
    if href:
        mary_oliver.append(href)

print("Number of links:", len(mary_oliver))

print(mary_oliver)
```

```python
#Problem 2

list_1 = [2, 2, 3, 3, 7, 15, 15, 21, 21, 21, 22, 22, 24, 30, 33, 38, 39, 42, 47, 49, 51, 51, 52, 53, 55, 58, 59, 60, 60, 61]
list_2 = [-19.26, 12.19, 9.29, -9.51, 14.80, 11.23, 3.63, -1.08, 16.16, 3.45, -2.77, 7.46, 7.38, 15.66, 14.73, 31.00, 26.48, 19.28, 14.42, 36.38, 33.35, 52.06, 28.63, 42.91, 32.10, 29.85, 14.40, 37.21, 21.56, 25.36]

import matplotlib.pyplot as plt

x_values = range(len(list_1))

plt.plot(x_values, list_1, label='List 1')
plt.plot(x_values, list_2, label='List 2')

plt.title('Line Graph')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()

plt.scatter(x_values, list_1, label='List 1', color='blue')
plt.scatter(x_values, list_2, label='List 2', color='red')


plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

x_values = range(len(list_1))
plt.scatter(x_values, list_1, label='List 1', color='blue')

slope, intercept, _, _, _ = linregress(x_values, list_1)
trend_line_y = slope * np.array(x_values) + intercept
plt.plot(x_values, trend_line_y, color='purple', label='Trend Line for List 1')

plt.title('Scatter Plot with Trend Line 1')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()

plt.scatter(x_values, list_2, label='List 2', color='red')

slope, intercept, _, _, _ = linregress(x_values, list_2)
trend_line_y = slope * np.array(x_values) + intercept
plt.plot(x_values, trend_line_y, color='green', label='Trend Line 2')

plt.title('Scatter Plot with Trend Line 2')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
#Problem 3

import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

data = {
    "TeamName": ["Reese", "Barnes", "Jones", "Javez", "Smith", "Heath", "Jacks", "Weeks", "Mitch", "Rankle"],
    "HighestScore": [112, 103, 127, 99, 130, 107, 89, 120, 122, 90],
    "LowestScore": [47, 87, 70, 51, 100, 101, 38, 97, 110, 55]
}

df = pd.DataFrame(data)

plt.scatter(df.index, df['HighestScore'], label='Highest Score', color='blue')
plt.scatter(df.index, df['LowestScore'], label='Lowest Score', color='red')

plt.title('Scatter Plot of Highest and Lowest Scores')
plt.xlabel('X')
plt.ylabel('Y')

plt.legend()

plt.show()
```
